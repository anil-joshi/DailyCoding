### Basics of Text Processing using NLTK

#### Natural Language Processing

Natural Language Processing is a field of Computer Science and Artificial Intelligence that helps gather meaningful insights from the natural language data used by human beings through the use of various techinques. As computers cannot understand and make sense of human data, the goal is to develop applications that can understand the words and sentences in the unstructured text to generate information from them. 



```python
# Load the necessary packages

import nltk
from nltk import word_tokenize
from nltk import sent_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Passing encoding = "ISO-8859-1" as I was having trouble loading the txt file.

file = open("Day 3/text.txt", encoding = "ISO-8859-1")
text = file.read()

print (text)
print(len(text))
```

```python
# Tokenizing the text by sentence and printing

sentence = sent_tokenize(text)
sentence
```

```python
# Tokenizing the text by word and printing

word = word_tokenize(text)
print(word)
```

```python
# Calculating the frequency of words

frequency = FreqDist(word)
frequency.most_common(20)
```

```python
# Removing the punctuation marks

# Create an empty list to store only words without punctuation marks
clean_words = []

# Go through the list of words and only keep alphabetical words
for w in word:
    if w.isalpha():
        clean_words.append(w.lower())
        
# Print the words without punctuation and check the length
print(clean_words)
print(len(clean_words))
```

```python
# Removing stopwords

# List of stopwords from the package
stop_words = stopwords.words('english')
print(stop_words)

# Create an empty list to store words only after removing the stop words
cleaner_words = []

# 
for c in clean_words:
    if c not in stop_words:
        cleaner_words.append(c)

# Print the words without stopwords and print the final length of words
print("\n")
print(cleaner_words)
print(len(cleaner_words))
```

```python
# Check the final frequency of words after cleaning

frequency = FreqDist(cleaner_words)
frequency.most_common(10)
```

```python
# Making a wordcloud

wordcloud = WordCloud().generate(text)
plt.figure(figsize = (10,10))
plt.axis("off")
plt.imshow(wordcloud)
plt.show()
```
